 Question 1: What is hypothesis testing in statistics?


- Hypothesis Testing is a statistical method used to make decisions or inferences about a population based on a sample of data.

- It helps to test whether a claim or assumption (hypothesis) about a population is true.

Steps in Hypothesis Testing:

1. Formulate Hypotheses:

   -  Null Hypothesis (H₀): Assumes no effect or no difference.

    Example: "The average exam score = 50."

   - Alternative Hypothesis (H₁): Assumes there is an effect or difference.

    Example: "The average exam score ≠ 50."

2. Choose Significance Level (α):

   - Commonly 0.05 (5%) → 5% chance of wrongly rejecting H₀.

3. Select Test and Compute Test Statistic:

 - e.g., t-test, z-test, chi-square test depending on data type.

4. Decision Rule:

   -  If p-value ≤ α → Reject H₀ (evidence supports H₁).

   - If p-value > α → Fail to reject H₀ (not enough evidence).

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

-  Null Hypothesis (H₀):  A default assumption that there is no effect, no difference, or no relationship in the population.

    - Example: “The average score of students = 50.”

- Alternative Hypothesis (H₁ or Ha):
A statement that contradicts the null hypothesis, claiming that there is an effect, difference, or relationship.

    - Example: “The average score of students ≠ 50.”
differences:

Null hypothesis :-
1. Assumes no effect/difference
2. Starting assumption to test
3. μ = 50
4. Retained unless evidence is strong

Alternative Hypothesis (H₁) :-  
1. Assumes effect/difference exists.
2. What researcher wants to prove
3. μ ≠ 50 (or μ > 50, μ < 50)
4. Accepted when H₀ is rejected

 Question 3 :  Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

-  Significance Level (α):

   - The probability of rejecting the null hypothesis when it is actually true (Type I error).

   - Common values: 0.05 (5%), 0.01 (1%).

- Role in Decision Making:

1. Compare the p-value with α.

   - If p-value ≤ α → Reject H₀ (evidence supports H₁).

    - If p-value > α → Fail to reject H₀ (not enough evidence).

2. Acts as a threshold for determining statistical significance.

Example:

If α = 0.05 and p-value = 0.03:

Since 0.03 ≤ 0.05 → Reject H₀.

- Conclusion: The result is statistically significant at 5% level.

Question 4: What are Type I and Type II errors? Give examples of each.

- Type I Error (False Positive): Rejecting the null hypothesis when it is actually true.
  - Example: A medical test concludes that a healthy patient has a disease (incorrect rejection).

- Type II Error (False Negative): Failing to reject the null hypothesis when it is actually false.
  - Example: A medical test concludes that a sick patient is healthy (failing to detect disease).

 In short:

Type I = "Detecting something that isn’t there."

Type II = "Missing something that actually exists."

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

- Z-test :
1. Used when the population variance (σ²) is known, or when the sample size is large (n > 30).

2. Assumes data is normally distributed.

3. Commonly used in quality control, large datasets.

- T-test:

1. Used when the population variance (σ²) is unknown and estimated from the sample.

2. More reliable with small sample sizes (n ≤ 30).

   - Examples include comparing the means of two groups (independent samples t-test) or before/after measurements (paired t-test).

* Rule of Thumb:

Large sample + known variance → Z-test

Small sample + unknown variance → T-test

Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10   # number of trials
p = 0.5  # probability of success
size = 1000  # number of samples

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(0, n+2), edgecolor='black', align='left')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()

#Interpretation:

# The distribution is centered around 5  (since expected value = n × p = 10 × 0.5 = 5).

# It is symmetric, resembling a bell curve (like Normal distribution), because p = 0.5.
```

Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python.

- We are testing whether the sample mean is equal to a hypothesized population mean (μ₀ = 50).

Steps:

  - Null Hypothesis (H₀): μ = 50

- Alternative Hypothesis (H₁): μ ≠ 50

- Significance Level (α): 0.05 (common choice)

- Compute Z-statistic = (x̄ − μ₀) / (σ/√n)

- Compare with Z-critical (±1.96 for α=0.05, two-tailed)

```python
import numpy as np
from math import sqrt
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu_0 = 50

# Calculate sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-test
z_stat = (sample_mean - mu_0) / (sample_std / sqrt(n))

# p-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Sample Std Dev:", sample_std)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the Null Hypothesis (H0).")
else:
    print("Fail to Reject the Null Hypothesis (H0).")
```

Interpretation:

- The sample mean is very close to 50.

- Z-statistic will be small (close to 0).

- p-value will be greater than 0.05, so we fail to reject H₀.

- Conclusion: There is no significant evidence that the mean differs from 50.

Question 8: Simulate normal distribution data & calculate 95% confidence interval for mean.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=100)  # mean=50, std=5, sample size=100

# Calculate sample mean and standard error
mean = np.mean(data)
sem = stats.sem(data)  # standard error of mean

# 95% Confidence Interval
confidence = 0.95
ci = stats.t.interval(confidence, len(data)-1, loc=mean, scale=sem)

print("Sample Mean:", mean)
print("95% Confidence Interval:", ci)

# Plot histogram of data
plt.hist(data, bins=15, color="skyblue", edgecolor="black", alpha=0.7)
plt.axvline(mean, color="red", linestyle="dashed", linewidth=2, label=f"Mean = {mean:.2f}")
plt.axvline(ci[0], color="green", linestyle="dashed", linewidth=2, label=f"Lower 95% CI = {ci[0]:.2f}")
plt.axvline(ci[1], color="green", linestyle="dashed", linewidth=2, label=f"Upper 95% CI = {ci[1]:.2f}")
plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

Interpretation:

- The sample mean will be close to 50.

- The 95% CI gives a range where the true population mean lies with 95% confidence.

Question 9: Calculate Z-scores and visualize standardized data.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def calculate_zscores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_scores = [(x - mean) / std for x in data]
    return np.array(z_scores)

# Example dataset
data = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Calculate Z-scores
z_scores = calculate_zscores(data)

print("Original Data:", data)
print("Z-scores:", z_scores)

# Plot histogram of Z-scores
plt.hist(z_scores, bins=10, color="orange", edgecolor="black", alpha=0.7)
plt.title("Histogram of Z-scores (Standardized Data)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.axvline(0, color="red", linestyle="dashed", linewidth=2, label="Mean (0)")
plt.legend()
plt.show()
```

Interpretation:

- A Z-score tells how many standard deviations a value is away from the mean.

   - Example: A Z-score of +2 means the value is 2 standard deviations above the mean; a Z-score of -1 means it is 1 standard deviation below the mean.

- Standardizing data (using Z-scores) helps compare values across different distributions.
